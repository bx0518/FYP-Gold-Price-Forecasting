# Retrieve data from Yahoo Finance

## Import Libraries

```python
import yfinance as yf
import numpy as np
import pandas as pd
```

## Download data of Gold & S&P500

```python
gold = yf.download("GC=F", start="2005-08-31", end="2024-09-01")
sp = yf.download("^SPX", start="2005-08-31", end="2024-09-01")
crudeoil = yf.download("CL=F", start="2005-08-31", end="2024-09-01")
usd = yf.download("DX-Y.NYB", start="2005-08-31", end="2024-09-01")
```

# Dataset Preparation

## Gold Dataset

```python
gold
```

## Gold + S&P500 Dataset

```python
gold_sp = gold.copy()
gold_sp['SP500'] = sp['Close']
gold_sp
```

## Gold + Crude Oil Dataset

```python
gold_crudeoil = gold.copy()
gold_crudeoil['Crude Oil'] = crudeoil['Close']
gold_crudeoil
```

## Gold + USD Index Dataset

```python
gold_usd = gold.copy()
gold_usd['USD Index'] = usd['Close']
gold_usd
```

## Gold + SP500 + Crude Oil + USD Index Dataset

```python
gold_three = gold.copy()
gold_three['SP500'] = sp['Close']
gold_three['Crude Oil'] = crudeoil['Close']
gold_three['USD Index'] = usd['Close']
gold_three
```

## Save the dataset into CSV file

```python
gold.to_csv("gold_price.csv")
gold_sp.to_csv("gold_price_sp500.csv")
gold_crudeoil.to_csv("gold_price_crude_oil.csv")
gold_usd.to_csv("gold_price_usd.csv")
gold_three.to_csv("gold_price_three.csv")
```
